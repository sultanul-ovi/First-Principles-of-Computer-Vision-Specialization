```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

from Implementation import *
```

```python
imagesDir = os.path.join("..\\..\\", "images", "Binary")
walk      = list(os.walk(imagesDir))
filenames = walk[0][2]
images = [file for file in filenames if ".png" in file]
multipleObjectImages = [filename for filename in images if "Multiple"   in filename]
multipleObjectImages
```

```python
for imageName in multipleObjectImages:
    imagePath = os.path.join(imagesDir, imageName)
    image     = cv2.imread(imagePath, cv2.IMREAD_UNCHANGED)
    
    numObjects, labeledImage, _, _       = getConnectedComponent(image)
    cv2NumObjects, cv2LabeledImage, _, _ = cv2.connectedComponentsWithStats(image, 1, cv2.CV_32S)
    
    plt.figure()
    plt.subplot(1, 2, 1)
    plt.title("#Objects: " + str(numObjects))
    plt.imshow(labeledImage)
    plt.subplot(1, 2, 2)
    plt.title("OpenCV #Objects: " + str(cv2NumObjects-1))
    plt.imshow(cv2LabeledImage)
```
